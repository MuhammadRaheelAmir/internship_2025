## Reading the Data

```python
## Reading a json file
import pandas as pd
from io import StringIO
Data = '{"Name": "Raheel", "email": "example@gmail.com", "Status": [{"title1":"Student", "title2":"Data Analyst"}]}'
df=pd.read_json(StringIO(Data))
df
```

```python
## Converting back to json
df.to_json() ## brings index as a key in each sub key:value pair
```

```python
df.to_json(orient='index') ## Only brings the index in the main key
```

```python
df.to_json(orient='records') ## removes the from key:value pair at all
```

```python
## Reading a csv file
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data")
df.head()
```

```python
## Converting back to csv
df.to_csv('wine.csv') ## Automatically saved
```
