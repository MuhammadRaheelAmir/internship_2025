```python
import seaborn as sns
```

```python
## Loading dataset with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
from matplotlib import pyplot as plt
```

```python
## Scatter plot with seaborn
## used to find the relationship
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter plot with total_bill vs tip')
plt.show()
```

```python
## Line plot with seaborn
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line plot of total_bill by size')
plt.show()
```

```python
## Bar plot with seaborn
## Used for comparison
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total_bill by day')
plt.show() 
```

```python
## Box plot with seaborn
## Used tocheck the composition of data
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('Box plot with total_bill by day')
plt.show()
```

```python
## Violin plot with seaborn
## Used to check the distribution of data
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violin plot of total_bill by day')
plt.show()
```

```python
## Histogram with seaborn
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.title('Histogram of total_bill')
plt.show()
```

```python
## KDE plot with seaborn
sns.kdeplot(tips['total_bill'],fill=True)
plt.title('KDE plot of total_bill')
plt.show()
```

```python
## Pair plot with seaborn
## Multiple plots together for better understanding
sns.pairplot(tips)
plt.show()
```

```python
## Correlation 
## Used to check the relation between entities
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
## Heatmap with seaborn
## Used to check Correlation between multiple entities
sns.heatmap(corr,annot=True,cmap='coolwarm') ## color-scale shows weak to strong correlation(blue to red)
plt.show()
```

```python
import pandas as pd 
data=pd.read_csv('sales_data.csv')
data.head(5)
```

```python
## Bar plot with total sales and product
sns.barplot(x='Product Category',y='Total Revenue',data=data,estimator=sum)
plt.title('Total Sales by Product Category')
plt.xlabel('Product')
plt.ylabel('Total sales')
plt.show()
```

```python
## Bar plot with total sales and region
sns.barplot(x='Region',y='Total Revenue',data=data,estimator=sum)
plt.title('Total sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```
